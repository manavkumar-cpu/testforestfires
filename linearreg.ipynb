```python
def first_fun(dest):
    with open(dest,'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=0
        char_count=0
        for line in lines:
            word=line.split()
            word_count+=len(word)
            for char in word:
                char_count+=len(char)
    print(f'lines={line_count},words={word_count},chars={char_count}')
first_fun('paste.txt')
                
```

```python
from multiprocessing.dummy import Process
import time
def squaree():
    for i in range(8):
        print(i*i)
        time.sleep(1)
        
def cubee():
    for i in range(8):
        print(i*i*i)
        time.sleep(1)
if __name__=='__main__':
    t1=Process(target=squaree)
    t2=Process(target=cubee)
    t=time.time()
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    t=time.time()-t
    print(f'time taken={t}')
```
